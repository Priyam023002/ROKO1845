```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from matplotlib.colors import ListedColormap
from sklearn.datasets import load_iris, fetch_california_housing, make_moons
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, mean_squared_error

# Set up plotting style
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
import kagglehub
path = kagglehub.dataset_download("uciml/iris")
```

```python
# 1. Load Data
iris = load_iris()
X_iris, y_iris = iris.data, iris.target

# 2. Split Data
X_train, X_test, y_train, y_test = train_test_split(
    X_iris, y_iris, test_size=0.3, random_state=42
)

# 3. **CRITICAL: Scale the Data**
scaler_c = StandardScaler()
X_train_scaled = scaler_c.fit_transform(X_train)
X_test_scaled = scaler_c.transform(X_test)

# 4. Create and Train (fit) the Model
# We'll pick K=5 for this example
knn_clf = KNeighborsClassifier(n_neighbors=5)
knn_clf.fit(X_train_scaled, y_train)

# 5. Make Predictions and Evaluate
y_pred = knn_clf.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
print(f"KNN Classifier (K=5) Accuracy: {accuracy * 100:.2f}%")
print("-" * 40)
```

```python
# 1. Load Data (using a subset to run faster)
housing = fetch_california_housing()
X_house, y_house = housing.data, housing.target
# Take a subset to make the example run quickly
X_subset, _, y_subset, _ = train_test_split(
    X_house, y_house, train_size=5000, random_state=42
)

# 2. Split Data
X_train_h, X_test_h, y_train_h, y_test_h = train_test_split(
    X_subset, y_subset, test_size=0.3, random_state=42
)

# 3. **CRITICAL: Scale the Data**
scaler_r = StandardScaler()
X_train_h_scaled = scaler_r.fit_transform(X_train_h)
X_test_h_scaled = scaler_r.transform(X_test_h)

# 4. Create and Train (fit) the Model
# We'll pick K=7 for this example
knn_reg = KNeighborsRegressor(n_neighbors=7)
knn_reg.fit(X_train_h_scaled, y_train_h)

# 5. Make Predictions and Evaluate
y_pred_h = knn_reg.predict(X_test_h_scaled)
mse = mean_squared_error(y_test_h, y_pred_h)
print(f"KNN Regressor (K=7) Root Mean Squared Error: {np.sqrt(mse):.4f}")
print("-" * 40)
```
