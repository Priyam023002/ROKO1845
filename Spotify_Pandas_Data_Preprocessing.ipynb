```python
import pandas as pd

df = pd.read_csv("light_spotify_dataset.csv")
df.head()
```

```python
value_missing = df.isnull().sum()

value_missing[value_missing>0]
```

```python
df_c=df.dropna(subset=['song']) #optimizing the query

df_c.isnull().sum()
df_c
```

```python
df_c.dtypes
```

```python
df_c['Release Date']=pd.to_datetime(df_c['Release Date'],format = '%Y')
```

```python
df_c
```

```python
# To identify the duplicate values
count = df_c.duplicated().sum()

print(count)

df_c.drop_duplicates()
```

```python
#TASK1: Average popularity by Emotion
#Find which emotions are associated

avg_emotion = df_c.groupby('emotion')['Popularity'].mean().sort_values(ascending=False)
print(avg_emotion)
```

```python
#Average Energy and Danceability by Genre

avg_genre = df_c.groupby('Genre')[['Energy', 'Danceability']].mean().sort_values(by=['Energy', 'Danceability'], ascending=[False,False])
print(avg_genre)
```

```python
# Top Artist by songs count
artist = df_c.groupby('artist')['song'].count().sort_values(ascending=False)
print(artist)
```

```python
loudest_artist = df_c.groupby('artist')['Loudness'].mean().sort_values(ascending=False)
print(loudest_artist)
```

```python
#How many songs released by an artist in a year
df_c.groupby(['artist','Release Date'])['song'].count().sort_values(ascending=False)
```
