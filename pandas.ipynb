```python
import numpy as np
np.random.seed(42)
outcomes = [1, 2, 3, 4, 5, 6]
probabilities = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]
random_variable = np.random.choice(outcomes, 1000, p=probabilities)
unique, counts = np.unique(random_variable, return_counts=True)
pmf = dict(zip(unique, counts / 1000))
print("Probability Mass Function (PMF):", pmf)
```
