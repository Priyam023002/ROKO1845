```python
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df)
plt.title("Total bill vs Tip")
plt.show()
```

```python
sns.barplot(data=df,x='day',y='total_bill',hue='time')
plt.show()
```

```python
sns.barplot(data=df, x = 'day', y ='total_bill', hue ='sex', estimator='median')
plt.title("Total bill vs Day")
plt.show()
```

```python
sns.barplot(data=df,x='day',y='time',hue='smoker')
plt.show()
```

```python
sns.barplot(data = df, x = 'time', y = 'tip', hue='sex')
plt.show()
```

```python
sns.barplot(data = df, x = 'time', y = 'tip', hue='sex')
plt.show()
```

```python
sns.histplot(data=df,x='total_bill',kde=True)
plt.show()
```

```python
sns.histplot(data = df, x = 'total_bill', hue = 'sex', multiple = 'stack')
plt.title("Total bill vs Day")
plt.show()
```

```python
df2 = sns.load_dataset('titanic')
sns.histplot(data = df2, x = 'age', kde = True, hue = 'class')
plt.show()
```

```python
df3=sns.load_dataset('diamonds')

sns.histplot(data=df3,x='price',hue='cut',kde=True,multiple ='stack')

plt.title("price Histogram")
plt.show()
```

```python
sns.violinplot(data=df3, x='cut', y='depth', palette='coolwarm')
plt.title("Price Distribution by Diamond Color")
plt.show()
```
